## Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("/Users/thaispalanca/Desktop/project macro/defnom.xlsx")
df.head()
```

```python
dates = df.set_index('data', inplace = True)
```

```python
defnom = df.loc[:,'defnom']
```

```python
defnom = pd.Series(defnom, index=dates)
```

```python
defplot = plt.figure(figsize=(10, 6))
plt.plot(defnom.index, defnom.values)
august = pd.Timestamp('2014-08-01')
plt.axvline(x=august, color='r', linestyle='--', label='August/2014')
plt.title('')
plt.xlabel('Years')
plt.ylabel('Nominal Deficit')
plt.legend()
defplot.savefig("./nominal deficit.png")
```
